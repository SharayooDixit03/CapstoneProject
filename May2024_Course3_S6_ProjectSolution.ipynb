```python
# Task 1
import random

riddles = {
    "What has one eye but cant see": "needle",
    "What has cities but no houses, forests but no trees, and water but no water": "Map",
    "What kind of lion never roars": "dandelion",
    "What do you call a bear with no teeth": "gummy bear",
    "I dont cry when you cut me, but you do. What am i": "onion",
    "What gets wet while its drying": "towel",
    "What has hands but cant clap?": "clock"
}

score = 0
questions = len(riddles)

while True:
    current_riddle, answer = random.choice(list(riddles.items()))
    print(f"\nRiddle: {current_riddle}")
    user_answer = input("Your answer: ").lower()
    
    if user_answer == answer.lower():
        score += 1
        print("Correct")
    else:
        print(f"Incorrect. The answer was: {answer}")
        
    play_again = input("Play again? (Y/N): ").upper()
    if play_again != "Y":
        break
percentage = (score/questions)*100
print(f"\nThanks for playing! Your Score: {percentage:.2f}%")
```

```python
# Task 2
import random

while True:
    num1 = random.randint(1, 20)
    num2 = random.randint(1, 20)
    operators = random.choice(['*', '-', '+', '/'])
    
    if operators == '/' and num2 == 0:
        continue
    
    x = f"{num1} {operators} {num2} ="
    your_answer = float(input(x))
    correct_answer = eval(f"{num1} {operators} {num2}")
    
    if your_answer == correct_answer:
        print("correct")
    else:
        print("incorrect. the correct answer is: ", correct_answer)
    
    play_again = input("play again? (y/n): ")
    if play_again.lower() != "y":
        break
```

```python
# task 3
import random

questions = {
    "Science": [
        {"question": "Which gas do plants absorb during photosynthesis?", "answer": "Carbon dioxide"},
        {"question": "What is the full name of the WWW?", "answer": "World Wide Web"},
        {"question": "How many bones are there in the adult human body?", "answer": "206"}
    ],
    "Mathematics": [
        {"question": "What is the only even prime number?", "answer": "2"},
        {"question": "What is the sum of the first 5 natural numbers?", "answer": "15"},
        {"question": "What is the name of the mathematical symbol π?", "answer": "Pi"}
    ],
    "Geography": [
        {"question": "What is the tallest mountain on Earth?", "answer": "Mount Everest"},
        {"question": "What is the capital of France?", "answer": "Paris"},
        {"question": "Which planet is known as the 'Red Planet'?", "answer": "Mars"}
    ],
    "Others": [
        {"question": "Which country won the most recent FIFA World Cup (as of 2022)?", "answer": "Argentina"},
        {"question": "Who invented the light bulb?", "answer": "Thomas Edison"},
        {"question": "In what year did India declare itself a Democratic and Republic state with the adoption of the Constitution?", "answer": "1950"}
    ]
}

def ask_category():
    print("Select a category:")
    for i, category in enumerate(questions.keys(), 1):
        print(f"{i}. {category}")
    choice = int(input("Enter the number of your choice: "))
    return list(questions.keys())[choice - 1]

def ask_question(question):
    print(question["question"])
    user_answer = input("Your answer: ")
    return user_answer

def check_answer(user_answer, correct_answer):
    if user_answer.lower() == correct_answer.lower():
        print("Correct!")
        return True
    else:
        print(f"Wrong! The correct answer is: {correct_answer}")
        return False

def display_score(score, total):
    print(f"Your score: {score}/{total} ({(score / total) * 100:.2f}%)")

def exec_questions(question_list):
    score = 0
    total = len(question_list)
    for question in question_list:
        user_answer = ask_question(question)
        if check_answer(user_answer, question["answer"]):
            score += 1
    display_score(score, total)

def play_game():
    print("Welcome to the General Knowledge Quiz!")
    while True:
        category = ask_category()
        print(f"\n{category} category:")
        exec_questions(questions[category])
        play_again = input("Do you want to play again? (y/n): ")
        if play_again.lower() != "y":
            break

if __name__ == "__main__":
    play_game()
    print("Thank you for playing!")
```
